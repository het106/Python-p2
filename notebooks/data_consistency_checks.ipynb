### Data Consistency Checks

```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
#
df = pd.read_csv("../data/publicmicrodatateachingsample.csv")
```

```python
# Display basic info about the data set
df.info()
```

```python
# Show the first few rows of the dataset
df.head()
```

```python
# Checking for missing values
print("\n Missing values per colomn:")
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check for duplicates
print("\nDuplicate rows in the dataset")
duplicates = df.duplicated().sum()
print(f"Number of duplicate nows: {duplicates}")
```

```python
# Check for inconsistent data types
```

```python
print("\nDate types of each colummn")
data_types = df.dtypes
print(data_types)
```

```python
# Check for outliers in the numerical columns
# Using inter quartile range

print("\nChecking for outliers in numerical columns using Inter Quartile Range")
numerical_columns = df.select_dtypes(include=[np.number]).columns

for col in numerical_columns:
    print(f"\n{col} column:")
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Setting the IQR bounds (if something is out of these bounds they are outliers)
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outliers = df[(df[col] < lower_bound) | (df[col] < upper_bound)]
    print(f"Number of outliers: {outliers.shape[0]}")

```

```python
# Report findings
print(f"Missing Values per Column:\n{missing_values}")
print(f"Duplicate Rows:\n{duplicates}")
```
